```python
%matplotlib inline
%reload_ext autoreload
%autoreload 2

import numpy as np
import pylab as plt
from vision.sim_tools.connectors.direction_connectors \
import direction_connection_angle_helper as dir_conn



ROW, COL = 11, 11
WIDTH, HEIGHT = 23, 23
MAX_ANGLE, MAX_DIST = 7, 6

def mapfunc(row, col, channel):
    return row*WIDTH + col

def extract_row_col(conns):
    rows = [c[0]//WIDTH for c in conns]
    cols = [c[0]%WIDTH for c in conns]
    
    return rows, cols
```

```python
img = np.zeros((WIDTH, HEIGHT, 3), dtype=np.uint8)

for dd in ['right2left', 'bl2tr', 'bottom2top', 'tl2br', \
           'left2right', 'br2tl', 'top2bottom', 'tr2bl']:
    conns = dir_conn(dd, MAX_ANGLE, MAX_DIST, COL, ROW, WIDTH, HEIGHT,
                     mapfunc, True, 2, 1)
    rows_E, cols_E = extract_row_col(conns[0])
    rows_I, cols_I = extract_row_col(conns[1])
    img[:] = 0
    img[rows_E, cols_E, 1] = 100
    img[rows_I, cols_I, 0] = 100
    
    fig = plt.figure()
    ax = plt.subplot(1, 1, 1)
    plt.imshow(img, interpolation='none')
    ax.set_title(dd)
    plt.show()
    
```
